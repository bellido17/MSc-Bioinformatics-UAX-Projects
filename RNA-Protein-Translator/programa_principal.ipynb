```python
import os
```

```python
ruta_fasta = "secuencia.fasta"
```

```python
if os.path.exists(ruta_fasta):
    print("El archivo existe.")
else:
    print("El archivo NO existe.")
```

```python
if os.path.isfile(ruta_fasta):
    print("Es un archivo válido.")
else:
    print("No es un archivo.")
```

```python
with open(ruta_fasta, "r") as archivo:
    contenido = archivo.read()
```

```python
print("Contenido del archivo FASTA:")
print(contenido)
```

```python
lineas = contenido.splitlines()
print(lineas)
```

```python
secuencias_arn = lineas[1]
print(secuencias_arn)
```

```python
secuencias_arn = ""

for linea in lineas:
    if not linea.startswith(">"):
        secuencias_arn += linea

print(secuencias_arn)
```

```python
import traductor
```

```python
proteina = traductor.traducir_a_proteinas(secuencias_arn)
print("Proteína traducida:")
print(proteina)
```

```python
with open("resultados_proteinas.txt", "wb") as f:
    f.write(proteina.encode("utf-8"))
```

```python
with open("resultados_proteinas.txt", "rb") as f:
    contenido_binario = f.read()

print("Contenido leído del archivo binario:")
print(contenido_binario)
```

```python
contenido_texto = contenido_binario.decode("utf-8")
print("Contenido decodificado:")
print(contenido_texto)
```
